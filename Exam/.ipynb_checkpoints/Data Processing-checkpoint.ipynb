>### The first thing on the list will be handling NaN values, and for this purpose I've got six posibilities to do it
1. NaN values column drop
> For the first option column droping is not the best option because we are going to lose a lot of important data the could be useful in the training porcess of the model
  
2. NaN Values row drop
> Row dropping is better than the first one but still we got some cases where we have over 450 NaN values so that would be about 15% of the data, which would be again, not the best idea
  
3. NaN values fill with previous entry
>Filling with the previous entry would be a better approach but when you get two whole differen cases with different results, this method might confuse the alghoritm so no to this one too
  
4. NaN values fill with median
<br>
5. NaN values fill with mean
> So because we have a normal distribution throughout each column of the dataset that means that the median and mean should be around the same spot, but still the median is more influenced by the outliers so because of that we are going to use the mean for this case


6. NaN values IterativeImputer
> Last and probably the most efficient method from our posibilites is the IterativeImputer that predicts the values from the other cases in the dataset, so we are definately going to use this one

```python
type(df_original)
```

>### First create the NaN mean replacement version 

```python
df_mean = df_original.copy()
for col in df_mean:
    df_mean[col].fillna((df_mean[col].mean()), inplace=True)

print(df_mean.isna().sum())
df_mean
```

>### And now the NaN iterative replacement version

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer(random_state = 3)

df_iterative = pd.DataFrame(imputer.fit_transform(df_original), columns = df_original.columns)

print(df_iterative.isna().sum())
df_iterative
```

>### As it is well known machine learning algorithms perform better when numerical input variables are scaled to a standard range, that's why we are going to go through this step too, and considering that StandardScaler is useful for the features that follow a Normal distribution, we are going to use it for this particular case

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
df_mean[df_mean.columns[:-1]] = pd.DataFrame(scaler.fit_transform(df_mean.drop(columns = ['Potability'])), columns = df_mean.columns[:-1])
df_mean
```

```python
df_iterative[df_iterative.columns[:-1]] = pd.DataFrame(scaler.fit_transform(df_iterative.drop(columns = ['Potability'])), columns = df_iterative.columns[:-1])
df_iterative
```

>### After this, it's time to deal with those outliers that were here and there all other the dataset, so even though we have a lot of methods to do that like:
>- Isolation Forest
>- One Class SVM
>- Standart Deviation
>- Local Outlier Factor
>- Eliptic Envelope
>### But for this time we are going to use the usual Isolation Forest method that gives good results in the majority of the cases

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(random_state = 3)
```

```python
df_mean['iso'] = model.fit_predict(df_mean.drop(columns = ['Potability']))

df_iterative['iso'] = model.fit_predict(df_iterative.drop(columns = ['Potability']))
```

```python
df_mean = (df_mean[df_mean.iso != -1]).reset_index(drop=True).drop(columns = ['iso'])

df_iterative = (df_iterative[df_iterative.iso != -1]).reset_index(drop=True).drop(columns = ['iso'])
```

>### Well according to the POINT BISERIAL SELECTOR all of the columns are important to the final results, therefore let's get to the models

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
def train(df, model):
    X = df.drop(columns=['Potability']).values
    y = df['Potability'].values

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 3)
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    accuracy=r2_score(y_test, y_pred)
    print("Accuracy:",accuracy,"%")
    
    cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
    class_names=[0,1] 
    fig, ax = plt.subplots()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names)
    plt.yticks(tick_marks, class_names)

    sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
    ax.xaxis.set_label_position("top")
    plt.tight_layout()
    plt.title('Confusion matrix', y=1.1)
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    
    return accuracy
```

```python
def train_with_data(model):
    train(df_mean, model)

    train(df_iterative, model)

    train(df_mean_smote_balanced, model)

    train(df_iterative_smote_balanced, model)
```

```python
train_with_data(LogisticRegression(random_state = 42))
```

```python
train_with_data(RandomForestClassifier(random_state = 42))
```

```python
train_with_data(KNeighborsClassifier())
```

```python
train_with_data(DecisionTreeClassifier(random_state = 42))
```

```python
train_with_data(GaussianNB())
```

>### An obvious conclusion from this results are that the smote balanced datasets performed way better compared to the simple ones

>### And the best performance was on the Random Forest Classifier with the mean smote balanced dataset, even though it has multiple wrong cases, it is the smallest number of positive negative predictions, which in a day to day life is very important to take into consideration

>### Lastly, let's try some Hyperparameter tunning

```python
from sklearn.model_selection import GridSearchCV
tuned_parameters = {
    'n_estimators':[a for a in np.arange(10, 150, 10)],
    'criterion': ["gini", "entropy"],
    "max_features": ["auto", "sqrt", "log2"]
}


X = df_mean_smote_balanced.drop(columns=['Potability']).values
y = df_mean_smote_balanced['Potability'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 3)

model = GridSearchCV(RandomForestClassifier(random_state = 42),  tuned_parameters)
model.fit(X_train, y_train)
print(model.best_params_)
```

```python
best_model = RandomForestClassifier(random_state = 42, criterion = "entropy", max_features = "auto", n_estimators = 90)
```

```python
train_with_data(best_model)
```
