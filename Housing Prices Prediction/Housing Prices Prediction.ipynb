```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Reading the data
train_data = pd.read_csv('train.csv', index_col = 'Id')
test_data = pd.read_csv('test.csv', index_col = 'Id')
```

```python
#Extracting the target and predictors
y = train_data.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = train_data[features].copy()
X_test = test_data[features].copy()
```

```python
#Splitting the data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 0)
```

```python
X_train.head()
```

```python
#Testing different types of models
from sklearn.ensemble import RandomForestRegressor

model_1 = RandomForestRegressor(n_estimators = 50, random_state = 0)

model_2 = RandomForestRegressor(n_estimators = 100, random_state = 0)

model_3 = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0)

model_4 = RandomForestRegressor(n_estimators = 200, min_samples_split =20, random_state = 0)

model_5 = RandomForestRegressor(n_estimators = 100, max_depth = 7, random_state = 0)

models = [model_1, model_2, model_3, model_4, model_5]
```

```python
#Analyzing the score of every created model
from sklearn.metrics import mean_absolute_error

#Creating a function for comparing different models
def score_model(model, X_t = X_train, X_v = X_valid, y_t = y_train, y_v = y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)

for i in range(0, len(models)):
    mae = score_model(models[i])
    print("Model %d MAE: %d" % (i+1, mae))
```

```python
#Choosing the best result
best_model = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0)

best_model.fit(X, y)

preds_test = best_model.predict(X_test)

output = pd.DataFrame({'Id':X_test.index, 'SalePrice': preds_test})

output.to_csv('Predictions.csv', index = False)

output
```
