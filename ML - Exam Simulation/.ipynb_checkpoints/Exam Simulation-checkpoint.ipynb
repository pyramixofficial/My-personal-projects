### VERY FUNNY

```python
def change(text):
    months = 0
    years = 0
    days = 0
    if type(text) is str:
        for i in text.split(sep=' ')[:-2]:
            if i.find('yr') !=-1:
                years = int(i[:i.find('y')])
            if i.find('m') !=-1:
                months = int(i[:i.find('m')])
            if i.find('d') !=-1:   
                days = int(i[:i.find('d')])
    return years * 12 +months + days/30
```

```python
change('8yrs 2mths 5days COE left)')
```

```python
data['Coe_left'] = data.Coe_left.apply(lambda x:change(x))
data
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer(random_state=1, max_iter=99)
```

```python
df = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
df
```

```python
df.info()
```

```python
from kydavra import PValueSelector

pselector = PValueSelector()
```

```python
columns = pselector.select(df, target='Price')
columns
```

```python
pselector.plot_process()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Price']).values
y = df['Price'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


regressor = LinearRegression()
regressor.fit(X_train, y_train)


y_pred=regressor.predict(X_test)
accuracy=r2_score(y_test, y_pred)
print("Accuracy:",accuracy,"%")
```
