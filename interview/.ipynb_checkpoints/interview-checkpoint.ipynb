```python
import pandas as pd
import numpy as np
import re
```

```python
with open('test.txt') as f:
    lines = f.readlines()

#Removing the "/n" from each line
for i in range(len(lines)):
    lines[i] = lines[i][:-1]

#Splitting the elements of the array
for i in range(len(lines)):
    lines[i] = lines[i].split(sep="  ")

#Removing some unnecessary characters from the string
for i in range(len(lines)):
    lines[i][0] = lines[i][0][:-1]

for i in range(len(lines)):
    index_date = lines[i][1].find(lines[i][0][0:2])
    lines[i].append(lines[i][1][index_date:index_date+19])

for i in range(len(lines)):
    lines[i][1] = lines[i][1].replace(lines[i][2], "")


def remove_dots(lines):
    return re.sub('\.+', '.', lines)


for i in range(len(lines)):
    lines[i][1] = remove_dots(lines[i][1])

for i in range(len(lines)):
     lines[i].append(lines[i][1][lines[i][1].find(".")+1:])
     lines[i][1] = lines[i][1].replace(lines[i][1][lines[i][1].find("."):], '')
lines
```

```python
df = pd.DataFrame(lines, columns = ["System_date_time", "Name", "Actual_date_time", "type"])
df[28:35]
```

```python
#Clearing the data
df = df.loc[(df["Name"] != "715[192") & (df["Name"] != "704[192")]
df = df.reset_index().drop("index", axis = 1)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df)
```

```python
#Creating a function for transforming the date in time to datetime object for better operations handling
from datetime import datetime

def transform_date(date_str): 
    return datetime.strptime(date_str, '%d/%m/%Y %H:%M:%S')

def transform_date2(date_str): 
    return datetime.strptime(date_str, '%d.%m.%Y %H:%M:%S')


df["System_date_time"] = df["System_date_time"].apply(transform_date)
df["Actual_date_time"] = df["Actual_date_time"].apply(transform_date2)

df["date"] = df["Actual_date_time"][0].date()
df["time"] = df["Actual_date_time"][0].time()
for i in range(len(df)):
    df["date"][i] = df["System_date_time"][i].date()
    df["time"][i] = df["System_date_time"][i].time()
df
```

```python
#Creating the result dataframe
data = pd.DataFrame(columns = ["Name", "Total_time", "Checkin_count", "Checkout_count", "Presence",
                       "Number_of_Days", "Average_checkin", "Average_checkout", "Average_time"
                       "Number_of_breaks", "Average_break"])
data
    
```

```python
data["Name"] = df.Name.unique()
data
```

#### Creating a dataset for only one person so I could easily make a function that will give me the values I need

```python
test = df.loc[df["Name"] == data["Name"][0]]
def get_values_from_user(info):
    #Calculating the number of checkouts
    nr_checkout = info["type"].value_counts()["CheckOut"]
    #Calculating the number of checkins
    nr_checkin = info["type"].value_counts()["CheckIn"]
    #Veryfying if the person is in the building
    presence = nr_checkin>nr_checkout
    
    #Defining the zero variables
    start = 0
    total_minutes = 0
    #Resetting the index
    info = info.reset_index().drop("index", axis = 1)
    #Getting the starting point in case the first case is a Checkout
    if info["type"][0] != "CheckIn":
        start = 1
    
    #Getting the total number of hours using datetime
    for i in range( ((len(info))//2)*2-1, start, -2):
        diff = info["Actual_date_time"][i] - info["Actual_date_time"][i-1]
        days, seconds = diff.days, diff.seconds
        total_minutes += days * 1440 + seconds // 60
    total_hours = total_minutes//60
    
    #Calculating number of days 
    nr_days = len(test["date"].value_counts())
    
    #Calculating average hours per day
    avg_hours = total_hours/nr_days
    
    return [nr_checkout, nr_checkin, presence, total_hours, nr_days, avg_hours]

get_values_from_user(test)
```
